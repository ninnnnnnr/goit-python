```python
python_data = python_data.groupby(['Должность'])
python_data
```

```python
new_df = python_data.agg({"Зарплата.в.месяц" : ["max", "min"]})
new_df
```

```python
import numpy as np


def fill_avg_salary(salary):
    return np.mean(salary)


df[["Зарплата.в.месяц"]].apply(fill_avg_salary)
```
