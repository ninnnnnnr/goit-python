```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
```

```python
A = 0.5
B = 0.3
N = 1000000
S0 = 990000
I0 = 7000
R0 = 3000
t0, tf = 0, 25
```

```python
def S(t, S0=S0, alpha=A):
    return S0 * np.exp(-alpha * t)
     

def I(t, I0=I0, S0=S0, alpha=A):
    return (I0 + alpha * S0 * t) * np.exp(-alpha * t)
     

def R(t, N=N):
    return N - S(t) - I(t)
```

```python
x = np.linspace(0, 20, 1000)
s = S(x)
i = I(x)
r = R(x)


plt.plot(x, s, label='S')
plt.plot(x, i, label="I")
plt.plot(x, r, label='R')
plt.legend(loc="right")
plt.show()
```

```python
minimize_scalar(lambda x, I0, S0, alpha: -I(x), bounds=(0, 20), args=(I0, S0, A))
```
