```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy import integrate
     
```

```python
time = np.linspace(0, 11, 12)
time
```

```python
speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
plt.scatter(time, speed)
plt.xlim(0,11)
plt.ylim(0,130)
plt.title('Значення точок швидкості')
plt.grid()
```

```python
speed_ = interp1d(time, speed, kind='cubic')
time_ = np.linspace(0, 11, 10000)
```

```python
plt.plot(time, speed, "o", time_, speed_(time_), '-')
plt.xlim((0, 11))
plt.ylim((0, 130))
plt.grid()
```

```python
res = integrate.quad(speed_, 0, 11)
res
```

```python
speed2_ = interp1d(time, speed, kind='quadratic')
```

```python
plt.plot(time, speed, "o", time_, speed2_(time_), '-')
plt.xlim((0, 11))
plt.ylim((0, 130))
plt.grid()
```

```python
res_quadratic = integrate.quad(lambda x: speed2_(x), 0, 11)
res_quadratic
```
