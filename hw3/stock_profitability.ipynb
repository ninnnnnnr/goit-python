```python
import pandas as pd
import seaborn as sns
```

```python
M = [1, 2, 3, 4, 5, 6]
A = [25, -10, 10, 5, 35, 13]
B = [0, 15, -5, 5, 20, 25]
C = [10, 25, -15, -5, -5, 15]

df = pd.DataFrame({'M':M, 'A': A, 'B': B, 'C': C})
```

```python
sns.regplot(x = "A",
            y = "B",
            data = df)
```

```python

sns.regplot(x = "C",
            y = "B",
            data = df, 
            color = 'green')
```

```python

sns.regplot(x = "A",
            y = "C",
            data = df, 
            color = 'red')
```
