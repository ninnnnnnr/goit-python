```python
import pandas as pd
```

```python
df = pd.read_html('https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C')


birth_rate = df[8]
```

```python
birth_rate = birth_rate.replace("—", None)
```

```python
birth_rate = birth_rate.astype({"2014": "float64", "2019": "float64"})
```

```python
birth_rate = birth_rate.drop([27], axis=0)
```

```python
birth_rate = birth_rate.fillna(birth_rate.mean(numeric_only=True))
```

```python
birth_rate[birth_rate["2019"]>birth_rate['2019'].mean()]["Регіон"]
```

```python
birth_rate.iloc[birth_rate['2014'].idxmax()]['Регіон']
```

```python
import matplotlib.pyplot as plt


plt.bar(birth_rate["Регіон"], birth_rate["2019"])
plt.xticks(rotation=90)
plt.show()
```
